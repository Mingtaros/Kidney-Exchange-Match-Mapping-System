```python
def read_pairs_file(pairs_filename):
    with open(pairs_filename, 'r') as f:
        pairs_data = f.read()

    pairs_data = pairs_data.split("\n")
    pairs_data = [np.array(data.split("\t")) for data in pairs_data][1:-1]
    
    return pd.DataFrame(pairs_data, columns=['pair_num', 'donor bloodtype', 'recipient bloodtype', 'pra'])
```
